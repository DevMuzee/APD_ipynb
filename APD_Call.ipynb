## What is the cumulative number of calls throughout each day, and how does this cumulative total change by sector?

```python
df.head(3)
```

## synatx transformation and representation

```python
## size number of calls throughout each day
cumsum_per_day=df.groupby(['sector','response_day_of_week'])['incident_number'].size().reset_index(name='count_no')
#cumsum by sector
cumsum_per_day['sector_wise']=cumsum_per_day.groupby('sector')['count_no'].cumsum()
#pivot table formatting
cumsum_per_day_pivot= cumsum_per_day.pivot_table(index='response_day_of_week', columns='sector', values='sector_wise', fill_value=0)

#resetting index
cumsum_per_day_pivot = cumsum_per_day_pivot.reset_index()

# Define mapping for numerical representation of days
day_mapping = {
    'Mon': 1, 'Tue': 2, 'Wed': 3, 'Thu': 4, 
    'Fri': 5, 'Sat': 6, 'Sun': 7
}

# Apply mapping to create a new column
cumsum_per_day_pivot['no_range'] = cumsum_per_day_pivot['response_day_of_week'].map(day_mapping)

# Sort by day of the week in ascending order
cumsum_per_day_pivot = cumsum_per_day_pivot.sort_values(by='no_range', ascending=True).reset_index(drop=True)

# Display the result
cumsum_per_day_pivot
```

## Graphical representation

```python
## size number of calls throughout each day
cumsum_per_day=df.groupby(['sector','response_day_of_week'])['incident_number'].size().reset_index(name='count_no')
#cumsum by sector
cumsum_per_day['sector_wise']=cumsum_per_day.groupby('sector')['count_no'].cumsum()
#pivot table formatting
cumsum_per_day_pivot= cumsum_per_day.pivot_table(index='response_day_of_week', columns='sector', values='sector_wise', fill_value=0)

#resetting index
cumsum_per_day_pivot = cumsum_per_day_pivot.reset_index()

# Define mapping for numerical representation of days
day_mapping = {
    'Mon': 1, 'Tue': 2, 'Wed': 3, 'Thu': 4, 
    'Fri': 5, 'Sat': 6, 'Sun': 7
}

# Apply mapping to create a new column
cumsum_per_day_pivot['no_range'] = cumsum_per_day_pivot['response_day_of_week'].map(day_mapping)

# Sort by day of the week in ascending order
cumsum_per_day_pivot = cumsum_per_day_pivot.sort_values(by='no_range', ascending=True).reset_index(drop=True)

# Melt the pivot table to long format for seaborn plotting
melted_df = pd.melt(cumsum_per_day_pivot, id_vars=['response_day_of_week'], value_vars=cumsum_per_day_pivot.columns[1:-1], var_name='sector', value_name='sector_wise')

# Plot using seaborn

plt.figure(figsize=(10, 6))
sns.lineplot(data=melted_df, x='response_day_of_week', y='sector_wise', hue='sector', marker='o')

plt.xlabel('Response Day of Week')
plt.ylabel('Cumulative Number of Calls')
plt.title('Cumulative Number of Calls Throughout Each Day by Sector')
plt.legend(title='Sector')
plt.grid(True)
plt.show()
```

## What are the most common types of incidents that occur between 10 PM and 6 AM?

```python
#Getting the time range between 10pm and 6am
start_hour=22
end_hour=6
time_range= df[(df['response_hour'] >= start_hour) | (df['response_hour']<= end_hour)]
#using time range to get value counts of the incident type
event_occuring=time_range['final_problem_category'].value_counts().reset_index(name='count_range')
#specifying columns needed
event_occuring.columns=['Incident_type',  'count_range']
#sorting to get the most recent
event_occuring.sort_values(by='count_range', ascending=False).head(5)
```
